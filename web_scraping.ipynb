```python
#perfomring a request and outputting the status code
#importing libs
from bs4 import BeautifulSoup
import requests

page = requests.get("https://www.geeksforgeeks.org/data-structures/")
print(page)
```

```python
#outputting the request response content
print(page.content)
```

```python
#turning our response into a beautifulsoup object to extract data
soup = BeautifulSoup(page.content, "html.parser")
print( soup.prettify() )
```

```python
#using find method to scrape text for first bold tag
title = soup.find("b")
print(title)
print( title.get_text() ) #returns all text within element
```

```python
#get all text within the bold element tag output
web_text = soup.find_all("b")
for text in web_text:
    print( text.get_text() )
```

```python
#getting elements by specific attribute key-values using github
page = requests.get("https://github.com/jomorro")
soup = BeautifulSoup(page.content, "html.parser")
username = soup.find( "span", attrs={ "class" : "vcard-username" }
)
print(username)
print( username.get_text() )
```

```python
#dom traversing - using the children attribute 
page = requests.get("https://www.geeksforgeeks.org/data-structures/")
soup = BeautifulSoup(page.content, "html.parser")
print(soup.children) #outputs an iterator object
```

```python
#create list of elements to loop over from object
for child in list( soup.children ):
    print( type( child ) )
```

```python
#getting tag object to access the title tag
html = list( soup.children )[2]
for section in html:
    print("\n\n Start of new section")
    print(section)
```
